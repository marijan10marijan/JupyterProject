### Load important libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Read csv file

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

### Numbers of rows and columns

```python
drinks.shape
```

### Data Type of each columns

```python
drinks.dtypes
```

### Numbers of Nan values in DF

```python
drinks.isna().sum()
```

### Group by continent 

```python
drinks.groupby('continent').count()
```

### Group by continent on a single column

```python
drinks.groupby('continent')['beer_servings'].count()
```

### Plot Beer servings data by continent

```python
data = drinks.groupby('continent')['beer_servings'].count()
plt.figure(figsize=(7,5))
plt.style.use('ggplot')
plt.plot(data)
plt.title('Beer servings by Continent')
plt.xlabel('Continent')
plt.ylabel('Beer servings')
plt.tight_layout()
```

### Bar chart avarage total_litres_of_pure_alcohol by continent 

```python
drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean().plot(kind='bar')
plt.ylabel('Litres')
plt.title('Avarage pure alcohol drinking')
```

### Sum, max, mean spirit servings for each continent

```python
drinks.groupby('continent')['spirit_servings'].agg(['sum','max','mean'])
```
